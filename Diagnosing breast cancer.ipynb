Now use a function from the sci-kit learn library to automatically split our data into training and test data. test_size tis the percentage of your data that your model will be tested on. random_state = 1 ensures that if the train_test_split function is called again the same data will be in the training data and same data will be in the test data.

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.15, random_state=1)
```

```python
def split_sizes(train, test):
    
    train_set_size = len(train)
    test_set_size = len(test)
    
    print(f'The size of the training set is: {train_set_size}')
    print(f'The size of the test set is: {test_set_size}')

split_sizes(train, test)
```

Now the features on the trainig set each have their own ranges. We want all the ranges to be the same otherwise the features with larger values may get favoured more. So it is common practice to normalise the features so that the values of each column range from 0 to 1. We can import the minmax scaler function from the sci-kit learn library to do this for us

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
train = pd.DataFrame(scaler.fit_transform(train), columns=train.columns)
```

```python
train
```

```python
test = pd.DataFrame(scaler.transform(test), columns=test.columns)
```

```python
test
```

fit_transform learns and applies the transformation to the data it's called on, while transform applies the same transformation to new data using the parameters learned from a previous fit_transform operation. This is crucial in maintaining the consistency of preprocessing across different datasets, especially when dealing with machine learning tasks to prevent data leakage and ensure a fair evaluation of model performance.

```python
prediction_var = ['perimeter_mean', 'compactness_mean', 'concavity_mean']
```

```python
train_x = train[prediction_var]
train_y = train.diagnosis

test_x = test[prediction_var]
test_y = test.diagnosis
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
model = MLPClassifier()
```

```python
model
```

Now train the model using our training data 

```python
model.fit(train_x, train_y)
```

```python
predictions = model.predict(test_x)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
def model_performance(y_true, y_pred):
    
    matrix = confusion_matrix(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)
    
    print(f"The confusion matrix is {matrix}")
    print("The precision score is %.2f" % precision)
    print("The recall score is %.2f" % recall)
    print("The accuracy score is %.2f" % accuracy)
```

```python
model_performance(test_y, predictions)
```
